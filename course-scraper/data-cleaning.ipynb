```python
import pandas as pd
import os
import numpy as np

pd.set_option("max_rows", 400)
pd.set_option("max_colwidth", 100)
```

### Read and select necessary columns

```python
desPath = "./data/DES-2021-courses-augmented.csv"
# print(os. getcwd())
raw = pd.read_csv(desPath, delimiter='\t')

# fields need cleaning: numberlinks, instructors, day_time
des = raw[["name", "numberlinks", "school", "major", "subject_code", "unit", "category", "target_year", "elecCompul",  "quarter"]].copy(deep=True) 
raw.head(10)

#des[des['subject_code']== "2313J"]
```

### Format numberlinks into comma/space separated list

```python
import re
#clean_numberlinks = lambda x: x[]
#des[numberlinks] = des[numberlinks].map()
def reformat_array_string(s):
#     s = '["' +s[1: -2] + ']'
    s = s[1: -2].replace(",", ' ')
    return re.sub(r'[\\|\"]', '', s)
    
s = des.iloc[0].numberlinks
des['numberlinks'] = des['numberlinks'].map(reformat_array_string)
# des['instructors'] = des['instructors'].map(reformat_array_string)

des.head(10)
# print(des.iloc[265].numberlinks[:-1] == ) 
```

### Explode dataframe by numberlinks


```python
des_cleaned = des.assign(numberlinks=des.numberlinks.str.split('[\s]+', )).explode('numberlinks') #.explode('numberlinks')
des_cleaned.iloc[2, 1] = "DES-DFT2713E"
des_cleaned.rename(columns={'name': 'subjectName'}, inplace=True)
des_cleaned.rename(columns={'subject_code': 'subjectCode'}, inplace=True)
des_cleaned.head(10)
# des_cleaned.numberlinks[des_cleaned.numberlinks.str.contains('DES') ]
# des_cleaned.head(10)
# pd.api.types.is_list_like(des['numberlinks'][0])
```

### Parse numberlinks and reset school, major, subjectCode

```python
# def setSchoolMajorSubjectCodeFromNumberLinks(row):     
#     print(row.loc["numberlinks"])
#     return row

def parseNumberlinks(el):
    pat = r"([A-Z]+)-([A-Z]{3})(\d+\w*)"
    m = re.match(pat, el)
    return [m.group(1), m.group(2), m.group(3)] if m else ['', '' , '']

smc = pd.DataFrame(pd.array(des_cleaned.numberlinks.map(parseNumberlinks)), columns=['school', 'major', 'subjectCode'])

des_cleaned.loc[:, ['school', 'major', 'subjectCode']] = smc.values
des_cleaned.head(20)
#des_cleaned.apply(setSchoolMajorSubjectCodeFromNumberLinks, axis=1) #, result_type='broadcast
```

### Fix rows whose school code's length are not 3

```python
des_cleaned.loc[des_cleaned.school.str.len() != 3, 'school'] =  'DES'
des_cleaned.loc[des_cleaned.school.str.len() != 3]
# school is always known, easy-piece
```

### Fix rows whose major code's length are not equal to 3

```python
des_cleaned[des_cleaned.major.str.len() != 3] 
```

### Fix rows whose major code's length are not equal to 5

```python
 des_cleaned.loc[des_cleaned.subjectCode.str.len() != 5, 'subjectCode'] = des_cleaned[des_cleaned.subjectCode.str.len() != 5].subjectCode.map(lambda val: val+'J').values
```

### courseMetaの唯一性を探る

```python
des_cleaned = des_cleaned.drop_duplicates(subset=['subjectName', "school", "major"])
output = des_cleaned.loc[:, des_cleaned.columns!='numberlinks']
# output[output.subjectName.str.contains("基礎造形Ⅰ")]
```

```python
output.to_csv('./DES-2021_cleaned.csv', index=False)
```
